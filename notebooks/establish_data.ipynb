## COLS Tables

````python
bet_info = pd.DataFrame({
    "Count": bets.count(),
    "Types": bets.dtypes,
    "Nulls": bets.isnull().sum()
}).reset_index()
bet_info
````

````python
games_info = pd.DataFrame({
    "Count": games.count(),
    "Types": games.dtypes,
    "Nulls": games.isnull().sum()
}).reset_index()
games_info
````

````python
lines_info = pd.DataFrame({
    "Count": lines.count(),
    "Types": lines.dtypes,
    "Nulls": lines.isnull().sum()
}).reset_index()
lines_info
````
